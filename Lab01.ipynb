###**Functions**
Functions in python are defined using the **def** keyword. See example code below:

**Additional Resources**
* <a href="https://www.youtube.com/watch?v=9Os0o3wzS_I&ab_channel=CoreySchafer">Corey Shafer Video</a>

```
# Example 1
# This defines the function "print_twice" which takes one argument
# (str_to_print). Running this cell will define the function (not call it, so
# there is no output)
def print_twice(str_to_print):
  print(str_to_print)
  print(str_to_print)
```

```
# This cell calls the function that we just defined.
print_twice("test string")
```

```
# Example 2 - Multiple arguments, optional arguments, return value
def multiply(first, second, third=1):
  # by setting 'third = 1', we have set a default value for the argument we
  # named 'third'. If only the first two arguments are specified, 'third' will
  # use the default value.
  return first * second * third

print("2 times 4 is: ", multiply(2, 4))
print("3 times 6 times 2 is: ", multiply(3, 6, 2))
```

###**For Loops**
for loops in python use the following syntax:

**Additional information on loops:**
* <a href="https://www.youtube.com/watch?v=6iF8Xb7Z3wQ&ab_channel=CoreySchafer"> Corey Shafer Video</a>
* <a href="https://wiki.python.org/moin/ForLoop"> Short written explanation</a>

```
for i in range(10):
  print(i)

# Note that range is a built in function which we use primarily for looping over
# a particular range of numbers (in this case 0-9). It is NOT a required part of
# the syntax of a for loop. If this seems confusing, you can read more here:
# https://www.geeksforgeeks.org/python-range-function/
```

The variable that we have named i in the above example is the variable that has its value changed every loop (in this case going from 0 to 9). Whatever we place after 'in' and before ':' must be something we can iterate over (an iterable). For example, this could be a list, or the output of the range function. See examples below.

```
# Example 1 - Looping through a list
my_list = ["a", 2, False]
for list_item in my_list:
  print(type(list_item))
```

```
# Example 2 - summing values from 1 to 10
total = 0
for i in range(1,11):
  total += i

print("Sum of numbers 1 to 10 is:", total)
```

In this section we will cover some of the basic functionality of matplotlib, a popular python library for making graphs / charts / figures.

**Additional Resources:**
* [Article on realpython.com](https://realpython.com/python-matplotlib-guide)  (same article as in Lab 0)

```
import matplotlib.pyplot as plt
```

We can create a simple plot very easily using the plot function, and passing it some x and y data.

```
x_data = [1,2,3,4]
y_data = [10,12,19,15]
plt.plot(x_data,y_data)
```

We could also make this a scatter plot:

```
plt.scatter(x_data,y_data)
```

When creating a plot, it's a good practice to create a variable for your figure and one for your axes so that we can refer to them. The image below indicates what parts of the chart that the figure and axes refer to. What isn't shown is that one figure could contain multiple charts (axes).

fig_map.webp 

```
# variables for our figure and axes objects
fig = plt.figure(figsize=(4, 4),)
# try changing the figsize (which is in inches) to say, (6,4)

ax = fig.add_subplot()

# notice that we are now calling the scatter method from the axes object
# where we want the scatter plot to appear
ax.scatter(x_data, y_data)
```

We can add various chart elements such as axis labels and a title.

```
fig = plt.figure(figsize=(4, 4),)
ax = fig.add_subplot()

ax.scatter(x_data, y_data)

#new elements
ax.set_xlabel('Horizontal Axis', fontsize=15)
ax.set_ylabel('Vertical Axis', fontsize=15)
ax.set_title('Meaningless Scatter Plot', fontsize=20)
```

In these most recent examples, we have used the object oriented approach to creating plots i.e., we have explicitly created figure and axes objects and then worked with them to get our desired plot.

The other way of creating plots is to call methods directly from plt e.g. plt.plot(x_data, y_data) as we did earlier, which can be more concise. In the next cell we will create the same plot as before with this method.

The applied portion of this course tends to use this second way of creating plots, but it is useful to be aware that there are two methods, and to think about all the components of matplotlib charts as being objects that can have their properties set - that is what's going on under the hood.

If you are confused about the distinction between these two methods, you can check out the article linked at the start of this section, and look under the heading "[Stateful Versus Stateless Approaches](https://realpython.com/python-matplotlib-guide/#stateful-versus-stateless-approaches)". Note that understanding this is not crucial to the course and you will be able to get by making charts without fully grasping it.

```
# Same chart as above, but only calling methods from plt directly
plt.figure(figsize=(4,4))
plt.scatter(x_data,y_data)
plt.xlabel('Horizontal Axis', fontsize=15)
plt.ylabel('Vertical Axis', fontsize=15)
plt.title('Meaningless Scatter Plot', fontsize=20);
# Asside: The semicolon at the end of the previous line is unimportant, but is
# suppressing uncaptured output.

# This method is doing the same things, but implicitly referring to our figure
# and axes, rather than having them specified as before.
```
